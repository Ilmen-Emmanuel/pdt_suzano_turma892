#### Questão 1

Criar um gráfico com a biblioteca Pandas que mostre a quantidade de pessoas por cidade de embarque, coluna embark_town no dataset.

```python
import pandas as pd
import seaborn as sns
```

```python
titanic = sns.load_dataset('titanic') # importando o dataset e guardando na variável
titanic.shape # verificando a quantidade de linhas e colunas do dataset
```

```python
titanic.head() # verificando a estrutura do dataset 
```

```python
titanic.info() # verificando os tipos de cada coluna 
```

```python
# contar quantas linhas para cada cidade existem
titanic['embark_town'].value_counts()
```

```python
# plotar um gráfico de barras com estas informações
titanic['embark_town'].value_counts().plot.bar(figsize= (16,8), title= 'Pessoas por cidade de embarque',xlabel='Cidade de embarque',ylabel='Quantidade de pessoas',color=['green','blue','red']);
```

#### Questão 2
Criar um gráfico com a biblioteca Pandas que mostre a quantidade de pessoas por cidade de embarque, coluna embark_town no dataset separados pelo sexo, coluna sex no dataset.

```python
pd.get_dummies(titanic, columns=['sex']) # retorna novas colunas com os tipos de dados encontrados na coluna 'sex
```

```python
# groupby filtra as cidades para as colunas definidas
pd.get_dummies(titanic, columns=['sex']).groupby('embark_town')[['sex_female','sex_male']].sum().plot.bar(figsize= (16,8), title= 'Pessoas por cidade de embarque separados por gênero',xlabel='Cidades de embarque',ylabel='Quantidade de pessoas');
```
